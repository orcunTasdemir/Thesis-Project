```python
def graph_folder(folder_name : str,
                 title : str = "title",
                 file_save :str = f"saveFigure{random.randint(1000,10000)}",
                 cycle_range : int = 20000,
                 variable : str = 'pop'):
    bin = []
    legends = []
    fig = plt.figure
    ax = plt.axes()
    for filename in os.listdir(f'data/{folder_name}/'):
        if "output" in filename:
            with open((f"data/{folder_name}/{filename}"), 'r') as f:
                data = json.load(f)
                var = []
                cycle = []
                for key in data:
                    if "header" == key:
                        continue
                    else:
                        var.append(data[key][variable])
                        cycle.append(data[key]['cycle'])
                ax.plot(cycle[:20000][::500], var[:20000][::500])
                legends.append(filename[7:-3])
    ax.set_title(f"{title}", fontsize='15')
    plt.legend(legends, bbox_to_anchor =(1.1, 1.1), ncol = 2)
    # plt.xticks([200, 500, 1000, 2000, 20000], rotation='vertical')
    plt.savefig(f'{file_save}', dpi=300, bbox_inches='tight')
graph_folder("SSS Single Run", "SSS populations", "sss_populations")
```
